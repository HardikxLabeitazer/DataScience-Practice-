```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
df = pd.read_csv("/content/drive/MyDrive/MachineLearning/LinearRegression/homeprices5hot.csv")
df
```

```python
dummies = pd.get_dummies(df.town)
dummies
```

```python
merged = pd.concat([df,dummies],axis="columns")
merged
```

```python
final = merged.drop(['town','west windsor'],axis="columns")
final
```

```python
model = LinearRegression()
```

```python
X = final.drop(['price'],axis="columns")
X
```

```python
Y = final.price
Y
```

```python
model.fit(X,Y)
```

```python
model.predict([[2800,0,1]])
```

```python
model.predict([[3400,0,0]])
```

Values for Predicting Windsor price


```python
model.score(X,Y)
```

```python
le = LabelEncoder()
```

```python
dfle = df
dfle.town =le.fit_transform(dfle.town)
dfle
```

```python
X = dfle[['town','area']].values
X
```

```python
Y = dfle.price
Y
```

```python
ohe = ColumnTransformer([("town",OneHotEncoder(),[0])],remainder='passthrough')
```

Using ColumnTransformer from sklearn to concentrate only on the main column which we are going to work with . In this case it is 'towns' column

```python
A =ohe.fit_transform(X)
A
```

```python
A = A[:,1:]
A
```

```python
model.fit(A,Y)
```

```python
model.predict([[1,0,2800]])
```

```python
model.predict([[0,1,3400]])
```

```python
carprices = pd.read_csv("/content/drive/MyDrive/MachineLearning/LinearRegression/carprices.csv")
carprices
```

```python
modelcar.predict([[86000,7,0,1]])
```
