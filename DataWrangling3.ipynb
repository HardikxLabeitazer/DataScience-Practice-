```python
df1 = pd.read_csv("/content/drive/MyDrive/DataAnalysis/earthquakes.csv",usecols=['time','title','place','magType','mag','alert','tsunami'])
```

```python
df1.head()
```

```python
df1['source']='USGS API'
df1.head()
```

```python
df1['mag_negative']=df1.mag<0
df1.head()
```

```python
df1.place.str.extract(r',(.*$)')[0].sort_values().unique()
```

```python
df1['parsed_place']=df1.place.str.replace(
    r'.* of','',regex=True 
).str.replace(
    'the',''
).str.replace(r'CA$','California',regex=True
).str.replace(
    r'NV$','Nevada',regex=True
).str.replace(
    r'MX$','Mexico',regex=True
).str.replace(
    r' region$','',regex=True
).str.replace(
    'northern',''
).str.replace(
    'Fiji Islands','Fiji'
).str.replace(
    r'^.*,','',regex=True
).str.strip()
```

```python
df1.parsed_place.sort_values().unique()
```

```python
df1.assign(
    in_ca = df1.parsed_place.str.endswith('California'),
    in_alaska = df1.parsed_place.str.endswith('Alaska'),
    neither=lambda x:~x.in_ca & ~x.in_alaska
).sample(5,random_state=0)
```

```python
tsunami = df1[df1.tsunami==1]
no_tsunami = df1[df1.tsunami==0]
```
